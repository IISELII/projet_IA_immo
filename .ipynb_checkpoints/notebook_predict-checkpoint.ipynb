```python
import io
import sys
import types
from nbformat import read
```

```python
def import_notebook(path):
    with io.open(path, 'r', encoding='utf-8') as f:
        nb = read(f, 4)

    ipynb_mod = types.ModuleType('notebook')
    sys.modules['notebook'] = ipynb_mod

    for cell in nb.cells:
        if cell.cell_type == 'code':
            exec(cell.source, ipynb_mod.__dict__)
            
    return ipynb_mod
```

# Split my data

```python
def df_split(df_ready):
    X = df_ready.drop(["median_house_value"], axis = 1)
    y = df_ready["median_house_value"]
    return X, y
```

```python
from sklearn.model_selection import train_test_split

def train_data(X, y, train_size = 0.8, random_state = 42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=train_size, random_state=random_state)
    return X_train, X_test, y_train, y_test
```

# Pipeline

```python
from sklearn.pipeline import Pipeline
```

```python
steps = [
    ('outlier', notebook_EDA.outlier_column(df_ready),
    'scaler', notebook_EDA.log_transformation(df_ready),
     'split', df_split(df_ready),
     'train', train_data(X, y, train_size = 0.8, random_state = 42),
    'classifier', RANSACRegressor(LinearRegression(),
max_trials=10, 
min_samples=2000, 
loss='absolute_loss', 
residual_threshold=10))
]
```

```python
pipeline = Pipeline(steps)
```

```python
pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_test)
```
